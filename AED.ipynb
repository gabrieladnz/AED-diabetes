```python
# Importações de bibliotecas
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
# carregando a base de dados com suas informações
df = pd.read_csv('Pokemon.csv', index_col=0, encoding='latin')
# Retorna as primeiras linhas do dataframe
df.head()
# print(df)
```

```python
# exibindo uma amostra dos dados
df.sample(10)
```

```python
# exibe as dimensões do array na forma de tupla (linhas)
df.shape
```

```python
# o método info() exibe uma rápida descrição dos dados, contendo número total de linhas, colunas e tipos.
df.info()
```

```python
# a função describe exibe estatísticas descritivas (contagem, média, máximo, mínimo...)
df.describe()
```

```python
# O método sns.lmplot() é usado para desenhar um gráfico de dispersão. Aqui estou analisando o ataque e a defesa.
sns.lmplot(x='Attack', y='Defense', data=df)
```

```python
# Criando gráfico de dispersão scatterplot com dois eixos: attack e defense
sns.lmplot(x='Attack', y='Defense', data=df,
           # traçando linha de regressão (não é exibida no gráfico)
           fit_reg=False,
           # adicionando cores
           hue='Stage')
```

```python
# Boxplot - Gráfico de caixa: ferramenta de visualização estatística

# Primeiro Quartil (Q1): A parte inferior da caixa representa o primeiro quartil, que é o valor que divide os 25% inferiores dos dados.
# Segundo Quartil (Q2): A linha dentro da caixa representa a mediana, que é o valor que divide o conjunto de dados ao meio (50% dos dados estão acima e 50% abaixo dessa linha). O segundo quartil é, portanto, igual à mediana.
# Terceiro Quartil (Q3): A parte superior da caixa representa o terceiro quartil, que é o valor que divide os 25% superiores dos dados.

plt.figure(figsize=(9,6))
sns.boxplot(data=df)
```

```python
# Criando um novo df a partir do df original (porém removendo algumas colunas)
stats_df = df.drop(['Total', 'Stage', 'Legendary'], axis=1)

# configurando dimensões do gráfico
plt.figure(figsize=(9,6))
# criando novo gráfico boxplot
sns.boxplot(data=stats_df)
```

```python
# Escolhendo tema
sns.set_style('whitegrid')

# Criando um gráfico Violin plot para visualizar a distribuição da variável nível de ataque em relação aos tipos de Pokémons.
plt.figure(figsize=(12,8)) # dimensões
sns.violinplot(x='Type 1', y='Attack', data=df) # criando gráfico
```

```python
pkmn_type_colors = ['#78C850',  # Grass
                    '#F08030',  # Fire
                    '#6890F0',  # Water
                    '#A8B820',  # Bug
                    '#A8A878',  # Normal
                    '#A040A0',  # Poison
                    '#F8D030',  # Electric
                    '#E0C068',  # Ground
                    '#EE99AC',  # Fairy
                    '#C03028',  # Fighting
                    '#F85888',  # Psychic
                    '#B8A038',  # Rock
                    '#705898',  # Ghost
                    '#98D8D8',  # Ice
                    '#7038F8',  # Dragon
                   ]
```

```python
# Gráfico com cores do tipo swarm plot (visualizando attack vs type)
plt.figure(figsize=(11,6)) # Set plot dimensions
sns.swarmplot(x='Type 1', y='Attack', data=df,
              palette=pkmn_type_colors)
```

```python
# Criando um gráfico violin plot e um gráfico swarm plot (relação attack vs type)
plt.figure(figsize=(11,6))

# Create plot
sns.violinplot(x='Type 1',
               y='Attack',
               data=df,
               inner=None, # Remove the bars inside the violins
               palette=pkmn_type_colors)

sns.swarmplot(x='Type 1',
              y='Attack',
              data=df,
              color='k', # Make points black
              alpha=0.7) # and slightly transparent

# Set title with matplotlib
plt.title('Attack by Type')
```

```python
# Criando distribuição conjunta (combina o scatter plot + histogramas). Esse gráfico é útil para ver a distribuição de cada variável separadamente
sns.jointplot(x='Attack', y='Defense', data=df)
```

```python
# Density Plotm - O gráfico de densidade é semelhante ao gráfico de dispersão, mas em vez de pontos individuais, ele usa uma representação suave da distribuição conjunta das variáveis. Ele ajuda a identificar áreas de alta densidade na distribuição dos dados.
sns.kdeplot(x='Attack', y='Defense', data=df)
```

```python
# Categorical Plot - verifica relações entre variáveis categóricas e numéricas
g = sns.catplot(x='Type 1',
                   y='Attack',
                   data=df,
                   hue='Stage',  # Color by stage
                   col='Stage',  # Separate by stage
                   kind='swarm') # Swarmplot

# Rotate x-axis labels
g.set_xticklabels(rotation=-45)
```

```python
# Count Plot (a.k.a. Bar Plot) - Visualiza a distribuição de variáveis categóricas (no caso type)
plt.figure(figsize=(9,6))
sns.countplot(x='Type 1', data=df, palette=pkmn_type_colors)

# Rotate x-labels
plt.xticks(rotation=-45)
```

```python
# Distribution Plot (a.k.a. Histogram) - Distribui a pontuação de Attack dos dados.
# A curva de densidade suaviza a distribuição e fornece uma estimativa visual da densidade de probabilidade subjacente. Isso pode ajudar a identificar tendências na distribuição dos valores de ataque.
plt.figure(figsize=(9,6))
sns.histplot(df, x='Attack',
             kde=True) # Plot the density curve too
```

```python
# Calculate correlations - Exibe a correlação entre todas as colunas numéricas e armazena na variável 'corr'
corr = df.corr()

# Heatmap
plt.figure(figsize=(9,8))
sns.heatmap(corr)
```

```python
# carregando a base de dados com suas informações
df_diabetes = pd.read_csv('diabetes.csv', index_col=0, encoding='latin')
# Retorna as primeiras linhas do dataframe
df_diabetes.head()
# print(df)
```

```python
# Amostragem dos dados - base diabetes.
df_diabetes.sample(10)
```

```python
# Número de elementos que representam a dimensão da base de dados - Diabetes
df_diabetes.shape
```

```python
# info() - exibe informações básicas sobre a base de dados (colunas, tipos, etc)
df_diabetes.info()
```

```python
# describe() - calcula um resumo das estatísticas do df_diabetes
df_diabetes.describe()
```

```python
# isnull().sum - retorna o número de valores ausentes no conjunto de dados
df_diabetes.isnull().sum
```

```python
# drop_duplicates() - retorna o DataFrame com as linhas duplicadas removidas.
duplicates = df_diabetes.drop_duplicates(subset="DiabetesPedigreeFunction")
duplicates
```

```python
# value_counts - retorna um objeto contendo contagens de valores únicos
df_diabetes.value_counts("DiabetesPedigreeFunction")
```

```python
# countplot() - usado para exibir a contagem de observações categóricas
sns.countplot(x='Age', data=df_diabetes, )
plt.show()
```

```python
# Comparação entre a glicose e o histórico familiar de diabetes por idade
sns.scatterplot(x='Glucose', y='DiabetesPedigreeFunction',
                hue='Age', data=df_diabetes, )

# Posicionando a legenda na figura
plt.legend(bbox_to_anchor=(1, 1), loc=2)

plt.show()
```

```python
# Comparação entre a glicose e a insulina por idade
sns.scatterplot(x='Glucose', y='Insulin',
                hue='Age', data=df_diabetes, )

# Posicionando a legenda na figura
plt.legend(bbox_to_anchor=(1, 1), loc=2)

plt.show()
```

```python
# Pairplot - Cria um conjunto de gráficos de dispersão (scatter plots) combinados com histogramas e representa as relações entre pares de variáveis no DataFrame df_diabetes
sns.pairplot(data=df_diabetes, hue='Age', height=2)
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10,10))

axes[0,0].set_title("Age")
axes[0,0].hist(df_diabetes['Age'], bins=7)

axes[0,1].set_title("DiabetesPedigreeFunction")
axes[0,1].hist(df_diabetes['DiabetesPedigreeFunction'], bins=5);

axes[1,0].set_title("Comprimento da Pétala")
axes[1,0].hist(df_diabetes['Comprimento da Pétala'], bins=6);

axes[1,1].set_title("Largura da Pétala")
axes[1,1].hist(df_diabetes['Largura da Pétala'], bins=6);
```

```python
# corr(method='pearson') - calcula a matriz de correlação entre as variáveis numéricas do DataFrame
df_diabetes.corr(method='pearson')
```

```python
# Heatmap - mapa de calor (verifica as correlações entre as variáveis)

# - cores mais escuras indicam correlações mais negativas
# - cores mais claras indicam correlações mais positivas
# - o tamanho e intensidade da cor indicam a magnitude da correlação.
sns.heatmap(df_diabetes.corr(method='pearson'),
            annot = True);

plt.show()
```
